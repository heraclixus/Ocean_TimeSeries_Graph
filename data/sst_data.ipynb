```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io
```

```python
data = scipy.io.loadmat("sst_pcs.mat")["pcs"]
```

```python
data.shape
```

```python
data[:,19]
```

```python
# 20 plots, each plotting time series from data
for i in range(19):
    fig = plt.figure()
    plt.plot(data[:,i])
    plt.title(f"PC {i+1}")
    plt.savefig(f"pc_{i+1}.png")
    plt.close(fig)
```

```python
# do the same but with subplots, each row 4 plots, 5 rows
fig, axs = plt.subplots(5, 4, figsize=(20, 20))
for i in range(20):
    axs[i//4, i%4].plot(data[:,i])
    axs[i//4, i%4].set_title(f"PC {i+1}")
plt.savefig("all_pcs.png")
plt.close(fig)

```
